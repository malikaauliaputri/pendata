###  3. Ambil contoh threshold split (misalnya):

* Threshold = 52 cm

Maka kita akan membagi:

* Group Left (≤ 52 cm)
* Group Right (> 52 cm)

###  4. Hitung Entropy Awal (sebelum split)

$$
Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

$$
p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}
$$

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15} = 0.996
$$

###  5. Entropy setelah split

Group Left (≤ 4.6 cm):

* Data: No 13, 9, 6, 15, 8, 7, 4, 10 → 8 data
* Kelas: 7 Mawar Cina, 1 Mawar Perancis

$$
Entropy(left) = -\frac{7}{8}\log_2\frac{7}{8} - \frac{1}{8}\log_2\frac{1}{8} \approx 0.543
$$

Group Right (> 4.6 cm):

* Data: No 12, 3, 1, 14, 2, 11, 5 → 7 data
* Kelas: 7 Mawar Perancis


$$
\text{Entropy(right)} = -\frac{7}{7} \log_2\left(\frac{7}{7}\right) = -1 \cdot \log_2(1) = 0
$$


Information Gain:

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot 0.543 + \frac{7}{15} \cdot 0 \right)
= 0.996 - (0.2896) = \boxed{0.7064}
$$

###  6. Gini Index

$$
Gini(D) = 1 - \left(\frac{8}{15}\right)^2 - \left(\frac{7}{15}\right)^2 = 1 - \frac{64}{225} - \frac{49}{225} = 1 - \frac{113}{225} = \frac{112}{225} \approx 0.497
$$

Group Left (≤ 4.6 cm):

* 7 Mawar cina, 1 Mawar perancis

$$
Gini(\text{left}) = 1 - \left(\frac{7}{8}\right)^2 - \left(\frac{1}{8}\right)^2 = 1 - 0.765625 - 0.015625 = 0.21875
$$

Group Right (> 265):

* 7 Mawar Perancis

$$
Gini(\text{right}) = 1 - \left(\frac{7}{7}\right)^2 = 1 - 1 = 0
$$

Gini Split:

$$
Gini_{\text{split}} = \frac{8}{15} \cdot 0.21875 + \frac{7}{15} \cdot 0 = \boxed{0.1167}
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Mawar ===
data = {
    "Panjang Tangkai (cm)": [55, 58, 54, 52, 57, 48, 50, 49, 47, 52, 60, 53, 46, 55, 48],
    "Diameter Bunga (cm)": [8.5, 8.2, 8.9, 7.8, 8.4, 6.5, 7.0, 6.8, 6.2, 7.4, 9.2, 8.5, 6.0, 8.2, 6.8],
    "Kelas": [
        "Rosa gallica", "Rosa gallica", "Rosa gallica", "Rosa gallica", "Rosa gallica",
        "Rosa chinensis", "Rosa chinensis", "Rosa chinensis", "Rosa chinensis", "Rosa chinensis",
        "Rosa gallica", "Rosa gallica", "Rosa chinensis", "Rosa gallica", "Rosa chinensis"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # gallica=0, chinensis=1

# === Fitur dan Target ===
X = df[["Panjang Tangkai (cm)", "Diameter Bunga (cm)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Cetak Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === 2. Visualisasi Pohon Keputusan ===
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=X.columns, class_names=le.classes_, filled=True, rounded=True)
plt.title("Decision Tree untuk Data Mawar")
plt.show()
```
